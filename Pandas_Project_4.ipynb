```python
# Display The Count of Paid and Free Courses

print(f"There are {(df['is_paid'] == True).sum()} Paid and {(df['is_paid'] == False).sum()} Free courses")
```

```python
# Which Course Has More Lectures (Free or Paid)?

df.head(2)
```

```python
df.groupby('is_paid')['num_lectures'].sum()
```

```python
# Which Courses Have A Higher Number of Subscribers Free or Paid?

free_courses=df[df['is_paid'] == False]['num_subscribers'].sum()
paid_courses=df[df['is_paid'] == True]['num_subscribers'].sum()

if free_courses > paid_courses:
    print("Free courses have more lectures.")
else:
    print("Paid courses have more lectures.")
```

```python
#  Which Level Has The Highest Number of Subscribers?

df.groupby('level')['num_subscribers'].sum()
```

```python
# Find Most Popular Course Title

df['course_title'].value_counts().sort_values(ascending=False).head(1)
```

```python
# Display 10 Most Popular Courses As Per Number of Subscribers

df.groupby('course_title')['num_subscribers'].mean().sort_values(ascending=False).head(10)
```

```python
# Find The Course Which Is Having The Highest Number of Reviews.

df.loc[df['num_reviews'].idxmax()]
```

```python
most_reviewed_course=df.loc[df['num_reviews'].idxmax()]

print(f"The course with the highest number of reviews is '{most_reviewed_course['course_title']}' with {most_reviewed_course['num_reviews']} reviews.")
```

```python
# Does Price Affect the Number of Reviews?

df[['price', 'num_reviews']].corr()
```

```python
# Find Total Number of Courses Related To Python

df.head(5)
```

```python
# Display 10 Most Popular Python Courses As Per Number of Subscribers

python_courses=df[df['course_title'].str.contains('Python', case=False, na=False)]

python_courses.groupby('course_title')['num_subscribers'].mean().sort_values(ascending=False).head(10)
```

```python
# In Which Year The Highest Number of Courses Were Posted?

df.head(2)
```

```python
df['year']=pd.to_datetime(df['published_timestamp']).dt.year
df.head()
```

```python
df.groupby('year')['course_id'].count().sort_values(ascending=False).head(1)
```

```python
# Display Category-Wise Count of Posted Subjects [Year Wise]

subject_year_count=df.groupby(['subject', 'year']).size().unstack()

subject_year_count.plot(kind='bar', figsize=(10,6), title='Category-Wise Count of Posted Subjects (Year Wise)')
```
