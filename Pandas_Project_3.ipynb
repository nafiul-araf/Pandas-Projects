```python
# Perform Data Cleaning [ Replace '?' with NaN ]

df.replace({'?': np.nan}, inplace=True)
df.head(10)
```

```python
# Drop all The Missing Values

df.dropna(inplace=True)
df.isnull().sum()
```

```python
# Check For Duplicate Data and Drop Them

duplicates=df.duplicated()
df[duplicates]
```

```python
len(df[duplicates])
```

```python
df.drop_duplicates(inplace=True)

len(df[df.duplicated()])
```

```python
# Drop The Columns education-num, capital-gain, and capital-loss

df.drop(['educational-num', 'capital-gain', 'capital-loss'], axis=1, inplace=True)
```

```python
# Replace Salary Values ['<=50K', '>50K'] With 0 and 1

df['income'].replace({'<=50K': 0, '>50K': 1}, inplace=True)
```

```python
# Which Workclass Getting The Highest Salary?

df.groupby('workclass')['income'].mean().sort_values(ascending=False).head(1)
```

```python
# How Has Better Chance To Get Salary greater than 50K Male or Female?

income_counts=df.groupby('gender')['income'].value_counts(normalize=True).unstack().fillna(0)
income_counts
```
