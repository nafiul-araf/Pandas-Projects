```python
salary=pd.read_csv('/content/drive/MyDrive/DS360 Machine Learning/Course DSML/Salaries.csv')
salary.head()
```

```python
salary.info()
```

```python
# what is the avg base pay

salary.BasePay.mean()
```

```python
# what is the highest amount of over time pay

round(salary.OvertimePay.max(), 2)
```

```python
# What is the job title of JOSEPH DRISCOLL? Note: Use all caps, otherwise you may get an answer that doesn't match up (there is also a lowercase Joseph Driscoll).

salary[salary['EmployeeName'] == 'JOSEPH DRISCOLL']['JobTitle']
```

```python
salary.Benefits.unique()
```

```python
# how much does JOSEPH DRISCOLL make (induding benefit)?

jd=salary[salary['EmployeeName'] == 'JOSEPH DRISCOLL']
jd[['BasePay', 'TotalPayBenefits']]
```

```python
# find the highest paid person with benefits

highest_paid_index=salary['BasePay'].idxmax()

person=salary.loc[highest_paid_index, 'EmployeeName']
benefit=salary.loc[highest_paid_index, 'TotalPayBenefits']

print(f"Highest pay {salary.BasePay.max()}\nThe highest-paid person is {person} with a benefit of {benefit}.")
```

```python
# what is the average base pay of all employees per year?

salary.groupby('Year')['BasePay'].mean()
```

```python
#How many people have the word Chief in their job title?


print(f"Number of people with 'Chief' in their job title: {salary[salary['JobTitle'].str.contains('Chief', case=False, na=False)].shape[0]}")
```
