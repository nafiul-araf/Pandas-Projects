```python
salary=pd.read_csv('/content/drive/MyDrive/DS360 Machine Learning/Course DSML/Salaries.csv')
salary.head()
```

```python
# Getting Information About Our Dataset Like Total Number Rows, Total Number of Columns, Datatypes of Each Column And Memory Requirement

salary.info()
```

```python
# Drop ID, Notes, Agency, and Status Columns

salary.drop(['Id', 'Notes', 'Agency', 'Status'], axis=1, inplace=True)
salary.head()
```

```python
# Get Overall Statistics About The Dataframe

salary.describe()
```

```python
# Find Occurrence of The Employee Names  (Top 5)

salary['EmployeeName'].value_counts().sort_values(ascending=False).head()
```

```python
# Find The Number of Unique Job Titles

len(salary['JobTitle'].unique())
```

```python
# Total Number of Job Titles Contain Captain

salary[salary['JobTitle'].str.contains('Captain', case=False, na=False)].shape[0]
```

```python
# Display All the Employee Names From Fire Department

salary[salary['JobTitle'].str.contains('Fire Department', case=False, na=False)]['EmployeeName']
```

```python
# what is the avg base pay

salary.BasePay.mean()
```

```python
# what is the highest amount of over time pay

round(salary.OvertimePay.max(), 2)
```

```python
# Replace 'Not Provided' in EmployeeName' Column to NaN

salary[salary['EmployeeName'] == 'Not Provided']['EmployeeName']
```

```python
salary['EmployeeName'].replace({'Not Provided': np.nan}, inplace=True)
```

```python
salary[salary['EmployeeName'] == 'Not Provided']['EmployeeName']
```

```python
salary.isnull().sum()
```

```python
# Drop The Rows Having 5 Missing Values

salary=salary[salary.isnull().sum(axis=1) != 5]
salary.isnull().sum()
```

```python
# What is the job title of JOSEPH DRISCOLL? Note: Use all caps, otherwise you may get an answer that doesn't match up (there is also a lowercase Joseph Driscoll).

salary[salary['EmployeeName'] == 'JOSEPH DRISCOLL']['JobTitle']
```

```python
salary.Benefits.unique()
```

```python
# how much does JOSEPH DRISCOLL make (induding benefit)?

jd=salary[salary['EmployeeName'] == 'JOSEPH DRISCOLL']
jd[['BasePay', 'TotalPayBenefits']]
```

```python
# find the highest paid person with benefits

highest_paid_index=salary['BasePay'].idxmax()

person=salary.loc[highest_paid_index, 'EmployeeName']
benefit=salary.loc[highest_paid_index, 'TotalPayBenefits']

print(f"Highest pay {salary.BasePay.max()}\nThe highest-paid person is {person} with a benefit of {benefit}.")
```

```python
# what is the average base pay of all employees per year?

salary.groupby('Year')['BasePay'].mean()
```

```python
# Find Average BasePay of Employee Having Job Title ACCOUNTANT

salary_acc=salary[salary['JobTitle'].str.contains('ACCOUNTANT', case=False, na=False)]

salary_acc['BasePay'].mean()
```

```python
# Find Average BasePay of All Employee Per JobTitle

salary.groupby('JobTitle')['BasePay'].mean()
```

```python
#How many people have the word Chief in their job title?


print(f"Number of people with 'Chief' in their job title: {salary[salary['JobTitle'].str.contains('Chief', case=False, na=False)].shape[0]}")
```

```python
# Find Top 5 Most Common Jobs

salary['JobTitle'].value_counts().sort_values(ascending=False).head()
```
