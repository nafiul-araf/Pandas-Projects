```python
# Data Cleaning  (Replace '--'  to NaN)

df.map(lambda x: '--' in str(x)).any()
```

```python
df['Rank']=df['Rank'].str.replace(r'[^\d]', '', regex=True).astype(int)
df
```

```python
# Data Cleaning [ Video Uploads & Subscribers ]

df['Video Uploads']=df['Video Uploads'].fillna(0).astype(int)
```

```python
df.isnull().sum()
```

```python
df['Subscribers']=df['Subscribers'].fillna(0).astype(int)
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

```python
# Data Cleaning [ Grade Column ]

df['Grade'].unique()
```

```python
df['Grade']=df['Grade'].str.strip()
df['Grade'].unique()
```

```python
df.isnull().sum()
```

```python
df=df[df['Grade'] != '']

df['Grade'].unique()
```

```python
df.head()
```

```python
# Find Average Views For Each Channel

df.groupby('Channel name')['Video views'].mean()
```

```python
# Find Out Top Five Channels With Maximum Number of Video Uploads

df.groupby('Channel name')['Video Uploads'].sum().sort_values(ascending=False).head()
```

```python
numeric_dtypes=df.select_dtypes(include='number')
print(numeric_dtypes.corr())
```

```python
# Which Grade Has A Maximum Number of Video Uploads?

most_uploaded_grade=df.loc[df['Video Uploads'].idxmax()]
```

```python
most_uploaded_grade
```

```python
most_uploaded_grade['Grade'], most_uploaded_grade['Video Uploads']
```

```python
# Which Grade Has The Highest Average Views?
grade_video_avg=df.groupby('Grade')['Video views'].mean()

grade_video_avg.idxmax(), grade_video_avg.max()
```

```python
# Which Grade Has The Highest Number of Subscribers?

most_subscribed_grade=df.loc[df['Subscribers'].idxmax()]

most_subscribed_grade['Grade'], most_subscribed_grade['Subscribers']
```

```python
# Which Grade Has The Highest Video Views?

most_viewed_grade=df.loc[df['Video views'].idxmax()]

most_viewed_grade['Grade'], most_viewed_grade['Video views']
```
